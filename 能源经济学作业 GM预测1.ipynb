```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.sans-serif"]="SimHei"
plt.rcParams["axes.unicode_minus"]=False
```

```python
gm=pd.read_csv("gm.csv",encoding="gbk")
gm
```

```python
#plt.plot([1,2,3,4,5],np.arange(2,20,4))
#没有标签从0开始
plt.xticks([1,2,3,4,5])
plt.yticks(np.arange(2,20,4))


plt.xlabel("Horizontal Axis")
plt.ylabel("Vertical Axis")
plt.title("Demonstration")
plt.show()
```

```python
plt.plot(gm.年份,gm.实际值,".")
plt.plot(gm.年份,gm.GM预测值)
plt.plot(gm.年份,gm.绝对误差)
```

```python
fig = plt.figure(figsize=(8,5),dpi=80)
#坐标轴1
ax1 = fig.add_subplot(111)
ax1.plot(gm.年份, gm.实际值,label='实际值',linewidth=3,)
ax1.plot(gm.年份, gm.GM预测值,color="orange",linestyle="--",linewidth=3,label='GM预测值')
ax1.legend(loc =8)
#坐标轴1标签
ax1.set_xlabel('年份')
ax1.set_ylabel('能源消费总量（万吨标煤）')
#图例位置
#plt.title('China Statistic Population')

#添加文本
#ax1.set_xticks(range(1960,2017))
#plt.setp(ax1.xaxis.get_majorticklabels(), rotation=90)
#坐标轴2

ax2 = ax1.twinx()

ax2.plot(gm.年份, gm.绝对误差,'gs',label='绝对误差')
ax2.set_ylabel('实际值与预测值之差')
ax2.axhline(0.0159072803052, color='grey', linestyle='--',label='0水平线') 
plt.grid()
ax2.legend(loc =9)
plt.show()
```

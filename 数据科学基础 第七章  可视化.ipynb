# 一.绘制带标签的sin和cos曲线

```python
#输入数据
X=np.linspace(-np.pi,np.pi,256,endpoint=True)
C,S=np.cos(X),np.sin(X)
matplotlib.rcParams['axes.unicode_minus']=False
#画图
plt.plot(X,C,color="blue",linewidth=3.0,linestyle="-",label="cos")
plt.plot(X,S,color="red",linewidth=3.0,linestyle="-",label="sin")
plt.legend(loc=2)
#标注刻度
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],
           [r"$-\pi$",r"$-\pi/2$",r"$0$",r"$+\pi/2$",r"$+\pi$"])
plt.yticks([-1,0,+1],[r"$-1$",r"$0$",r"$+1$"])
#取消两个坐标轴
ax=plt.gca()
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
#移动坐标轴
ax.xaxis.set_ticks_position("bottom")
ax.spines["bottom"].set_position(("data",0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data",0))
t=2*np.pi/3
#绘制线
plt.plot([t,t],[0,np.cos(t)],color="blue",linewidth=2.5,linestyle="--")
plt.scatter([t,],[np.cos(t),],50,color="blue")
#绘制标签
plt.annotate(r"$\sin\left<\frac{2\pi}{3}\right)=\frac{\sqrt{3}}{2}$",
            xy=(t,np.sin(t)),xytext=(1.5,1.1),fontsize=16,
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=-0.5"))
plt.plot([t,t],[0,np.sin(t)],color="red",linewidth=2.5,linestyle="--")
plt.scatter([t,],[np.sin(t),],50,color="red")
plt.annotate(r"$\cos\left<\frac{2\pi}{3}\right)=-\frac{\sqrt{3}}{2}$",
            xy=(t,np.cos(t)),xytext=(1.5,-1.1),fontsize=16,
            arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=-0.5"))
plt.plot([t,t],[0,np.sin(t)],color="red",linewidth=2.5,linestyle="--")
plt.show
```

# 二.绘制北京，上海，家乡(江西)的城市各性别人口箱线图

```python
china=pd.read_csv("renkou18.csv",encoding="gbk")
plt.rc("font",family="SimHei",size=13)
#处理数据
beijing=pd.pivot_table(china,values="北京",index=["年龄"],columns=["性别"])
shanghai=pd.pivot_table(china,values="上海",index=["年龄"],columns=["性别"])
jiangxi=pd.pivot_table(china,values="江西",index=["年龄"],columns=["性别"])
dff=pd.concat([beijing,shanghai,jiangxi],axis=1)
#绘制箱线图
dff.boxplot()
plt.xticks([1,2,3,4,5,6,7,8,9],[" ","北京"," "," ","天津"," "," ","江西"," "])
for i in np.arange(3):
    plt.annotate(s="女",xy=(3*i+1,30),xytext=(-5,30),textcoords="offset points")
for i in np.arange(3):
    plt.annotate(s="合计",xy=(3*i+2,30),xytext=(-5,30),textcoords="offset points")
for i in np.arange(3):
    plt.annotate(s="男",xy=(3*i+3,30),xytext=(-5,30),textcoords="offset points")
plt.xlabel("城市")
plt.ylabel("人口（千万）")
plt.title("北京，天津，江西各性别人口箱线图")
```

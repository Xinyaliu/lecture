```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
 
plt.rcParams["font.sans-serif"]="SimHei"
plt.rcParams["axes.unicode_minus"]=False
```

```python
dm=pd.read_excel("终端消费.xlsx",encoding = 'GB2312')
dm1=dm[dm.省份=="北京"]
dm2=dm[dm.省份=="天津" ]
dm3=dm[dm.省份=="河北" ]
dm4=dm[dm.省份=="广东" ]
```

```python
plt.figure(figsize=(7,4),dpi=200)
N = 19
S =  dm1.第一产业占比
C = dm1.第二产业占比
M = dm1.第三产业占比 
D=S+C+M
K= dm1.年份 
d=[]

for i in range(0,len(S)):
    sum = S[i] + C[i]
    d.append(sum)

    
#menStd = (2, 3, 4, 1, 2)
#womenStd = (3, 5, 2, 3, 3)
ind = np.arange(N)    # the x locations for the groups
width = 0.8       # the width of the bars: can also be len(x) sequence
 
p1 = plt.bar(ind, S/D, width, color="springgreen",edgecolor='black',linewidth=0.7)#, yerr=menStd)
p2 = plt.bar(ind, C/D, width, bottom=S/D, color="cornflowerblue",edgecolor='black',linewidth=0.7)#, yerr=womenStd)
p3 = plt.bar(ind, M/D, width, bottom=d/D, color="lightsalmon",edgecolor='black',linewidth=0.7)
 
plt.ylabel('百分比')
#plt.title('Scores by group and gender')
plt.xticks(ind, K)
plt.yticks(size=12)
plt.xticks(rotation=90,size=10)

plt.legend((p1[0], p2[0], p3[0]), ('第一产业', '第二产业', '第三产业'))
plt.savefig('北京市产业结构.jpg')
```
